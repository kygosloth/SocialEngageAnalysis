```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

```

```python

categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']

n = 500
data = {
    'Date': pd.date_range('2021-01-01', periods=n),
    'Category': [random.choice(categories) for _ in range(n)],
    'Likes': np.random.randint(0, 10000, size=n)
}
```

```python
# Create DataFrame
df = pd.DataFrame(data)

# Print head, info, and description
print(df.head())
print(df.info())
print(df.describe())

# Count occurrences of each category
print(df['Category'].value_counts())
```

```python
# Remove null values
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()

# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Convert 'Likes' to integers
df['Likes'] = df['Likes'].astype(int)
```

```python
# Visualize data
sns.histplot(data=df, x='Likes')
plt.show()

sns.boxplot(data=df, x='Category', y='Likes')
plt.show()

# Analyze data
print("Mean Likes:", df['Likes'].mean())
print("Mean Likes per Category:")
print(df.groupby('Category')['Likes'].mean())

```
